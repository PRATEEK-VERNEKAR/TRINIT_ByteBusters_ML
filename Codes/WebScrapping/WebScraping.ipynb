```python
desc = []
desc.clear()
```

```python
for i in range(20):
    if(len(tweets[i]['tweets'])!=0):
       desc.append(tweets[i]['tweets'])
print(desc)
```

```python
data = {
'link' : [],
'text':[],
'date':[],
'name' :[] , 
}

for i in range(5):
    for j in range(len(desc[i])):
        data['link'].append(desc[i][j]['link']) 
        data['text'].append(desc[i][j]['text']) 
        data['date'].append(desc[i][j]['date']) 
        data['name'].append(desc[i][j]['user']['name'])
        
        
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df.head()
```
